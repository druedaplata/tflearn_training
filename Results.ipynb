```python
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(style="darkgrid")
```

# Starlight Dataset

```python
train = pd.read_csv('data/StarLightCurves/StarLightCurves_TRAIN', header=None)
test = pd.read_csv('data/StarLightCurves/StarLightCurves_TEST', header=None)

fig, ax = plt.subplots(1, 2)
fig.set_size_inches(10,5)
sns.countplot(x=0, data=train, ax=ax[0])
ax[0].set_title('Train')
sns.countplot(x=0, data=test, ax=ax[1])
ax[1].set_title('Test')
plt.show()
```

# Accuracy

```python
vgg_acc_val = pd.read_csv('csv/starlight/run_star_vgg-tag-Accuracy_Validation.csv')
cnn_acc_val = pd.read_csv('csv/starlight/run_star_cnn-tag-Accuracy_Validation.csv')
lstm_acc_val = pd.read_csv('csv/starlight/run_star_lstm-tag-Accuracy_Validation.csv')

fig, ax = plt.subplots(1, 1)
fig.set_size_inches(15,5)
ax.plot(vgg_acc_val['Value'], 'r', label='vgg')
ax.plot(cnn_acc_val['Value'], 'g', label='alexnet')
ax.plot(lstm_acc_val['Value'], 'b', label='lstm')
ax.set_title('Validation Accuracy')
ax.legend()


plt.show()
```

# Loss

```python
vgg_loss_train = pd.read_csv('csv/starlight/run_star_vgg-tag-Loss.csv')
vgg_loss_val = pd.read_csv('csv/starlight/run_star_vgg-tag-Loss_Validation.csv')

cnn_loss_train = pd.read_csv('csv/starlight/run_star_cnn-tag-Loss.csv')
cnn_loss_val = pd.read_csv('csv/starlight/run_star_cnn-tag-Loss_Validation.csv')

lstm_loss_train = pd.read_csv('csv/starlight/run_star_lstm-tag-Loss.csv')
lstm_loss_val = pd.read_csv('csv/starlight/run_star_lstm-tag-Loss_Validation.csv')

fig, ax = plt.subplots(3, 2)
fig.set_size_inches(15,15)

ax[0][0].plot(vgg_loss_train['Value'], 'r')
ax[0][1].plot(vgg_loss_val['Value'], 'g')
ax[0][0].set_title('VGG Finetune - Loss Train')
ax[0][1].set_title('VGG Finetune - Loss Val')

ax[1][0].plot(cnn_loss_train['Value'], 'r')
ax[1][1].plot(cnn_loss_val['Value'], 'g')
ax[1][0].set_title('Alexnet - Loss Train')
ax[1][1].set_title('Alexnet - Loss Val')

ax[2][0].plot(lstm_loss_train['Value'], 'r')
ax[2][1].plot(lstm_loss_val['Value'], 'g')
ax[2][0].set_title('LSTM - Loss Train')
ax[2][1].set_title('LSTM - Loss Val')
plt.show()
```
